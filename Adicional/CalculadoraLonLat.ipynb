```python
import pandas as pd
import requests

# Configurar credenciales de la API de Bing Maps
bing_maps_api_key = "llenar este esapcio con clave API de Bing Maps"
base_url = "https://dev.virtualearth.net/REST/v1/Locations"

# Leer la hoja de Excel en un dataframe, en la hoja se tienen dos columnas una de ID SERVICIO y otra de DIRECCION
datos_df = pd.read_excel("DatosLonLat.xlsx", sheet_name="Datos", usecols="A,B")

# Crear columnas para la longitud y latitud
datos_df["LONGITUD"] = None
datos_df["LATITUD"] = None

#Encontrar la longitud y latitud de cada dirección y guardarla en el dataframe
for index, row in datos_df.iterrows():
    direccion = row["DIRECCION"]
    querystring = {"q": direccion, "key": bing_maps_api_key}
    response = requests.request("GET", base_url, params=querystring)
    json_response = response.json()
    if json_response["resourceSets"][0]["estimatedTotal"] > 0:
        latitud = json_response["resourceSets"][0]["resources"][0]["point"]["coordinates"][0]
        longitud = json_response["resourceSets"][0]["resources"][0]["point"]["coordinates"][1]
        datos_df.at[index, "LONGITUD"] = longitud
        datos_df.at[index, "LATITUD"] = latitud

# Mostrar el dataframe con las nuevas columnas
datos_df

# Se imprime la longitud y latitud de cada dirección en el archivo Excel en las columnas C y D
datos_df.to_excel("DatosLonLat_Resultados_1.xlsx", sheet_name="Resultados", index=False)
```

```python
import numpy as np
import folium

# Crear un arreglo aleatorio de frecuencias
frecuencias = np.random.normal(loc=100, scale=10, size=(100,))

# Crear un arreglo aleatorio de ubicaciones (latitud, longitud)
ubicaciones = np.random.uniform(low=[4.490, -74.260], high=[4.870, -73.970], size=(100, 2))

# Crear el mapa centrado en Bogotá
mapa = folium.Map(location=[4.6097, -74.0817], zoom_start=11)

# Iterar sobre las ubicaciones y agregar marcadores al mapa
for i, ubicacion in enumerate(ubicaciones):
    folium.Marker(ubicacion, popup=frecuencias[i]).add_to(mapa)

# Mostrar el mapa
mapa
```
